```python
import numpy as np
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading dataset
dataframe = pd.read_csv("./predictorsCSV.csv")
normalizedDF = preprocessing.normalize(dataframe)
```

```python
granite=normalizedDF[:,0]
plt.hist(granite,  bins=30)
plt.ylabel('Frequency');
plt.xlabel('Distance to Granite');
plt.xlim((-1,1))
```

```python
fault=normalizedDF[:,1]
plt.hist(fault,  bins=30)
plt.ylabel('Frequency');
plt.xlabel('Distance to Fault');
plt.xlim((-1,1))
```

```python
copper=normalizedDF[:,2]
plt.scatter(copper, copper)
plt.xlabel('Copper PPM')
plt.ylabel('Copper PPM')
plt.xlim((-1,1))
plt.show()
```

```python
iron=normalizedDF[:,3]
plt.scatter(copper, copper)
plt.xlabel('Iron PPM')
plt.ylabel('Iron PPM')
plt.xlim((-1,1))
plt.show()
```
